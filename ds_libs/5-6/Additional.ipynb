```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
```

```python
data = load_wine()
```

```python
data_keys = data.feature_names
data_keys
```

```python
X = pd.DataFrame(data.data, columns=data_keys)
X.head(n=10)
```

```python
X.shape
```

```python
X.info()
```

```python
target = pd.DataFrame(data.target, columns=['target'], dtype=np.int64)
target
```

```python
X = pd.concat([X, target], axis=1)
X
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
X_corr = X.corr()
X_corr

fig, axes = plt.subplots(1, 1, figsize=(12, 12))
axes.set_title('Correlation matrix', fontsize=15)
sns.heatmap(X_corr, annot=True, linewidth=1)
```

```python
high_corr = []
X_corr_target = X.corrwith(X['target'])
# -2 to drop 'target' index
i = len(X_corr_target) - 2 
while i > 0: 
    if abs(X_corr_target[i]) > 0.5:
        high_corr.append(X_corr_target.index[i])
    i -= 1
high_corr
```

```python
X.drop('target', axis=1)
```

```python
X_high_corr = pd.DataFrame(X, columns=high_corr)
X_high_corr_pow = X_high_corr.pow(2)
columns_pow = []
for column in list(X_high_corr_pow):
    columns_pow.append(f'{column}_2')
X_high_corr_pow = X_high_corr_pow.set_axis(columns_pow, axis=1)
X_high_corr_pow
```

```python
X = pd.concat([X, X_high_corr_pow], axis=1)
```
