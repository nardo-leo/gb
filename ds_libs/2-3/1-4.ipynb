```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
DATASET_PATH = '~/Downloads/creditcard.csv'
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv(DATASET_PATH)
df.tail(n=2)
```

```python
df_class = df['Class'].value_counts(normalize=True)
```

```python
df_class.plot(kind='bar')
plt.show()
```

```python
df_class.plot(kind='bar', logy=True)
plt.show()
```

```python
df_v1_crime = df['V1'].loc[df['Class'] == 1]
plt.hist(df_v1_crime, density=True, bins=20, alpha=0.5, color='red')
plt.xlabel('Class')
plt.legend(labels='1', loc='upper left')
plt.show()
```

```python
df_v1_reg = df['V1'].loc[df['Class'] == 0]
plt.hist(df_v1_reg, density=True, bins=20, alpha=0.5, color='grey')
plt.xlabel('Class')
plt.legend(labels='0', loc='upper left')
plt.show()
```
